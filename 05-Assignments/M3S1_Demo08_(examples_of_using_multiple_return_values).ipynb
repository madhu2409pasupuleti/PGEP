Imagine you are a contractor who is responsible for painting the outsides of a group of identical suburban houses. The houses can be thought of as cuboids. Let's write a program which would output the surface area each house, as well as the cost to paint the houses once the user inputs the dimensions of the house, the price per square metre of paint, and the number of houses.

```python
def compute_area_cost(length, width, height, price_per_sq_metre, n_houses):
    surface_area = 2 * (length + width) * height + length * width  # total surface area excluding the floor
    cost_to_paint = surface_area * price_per_sq_metre * n_houses  # cost to paint the total surface area

    return surface_area * n_houses, cost_to_paint
```

Let's test our function

```python
length = float(input('Enter the length of the house in metres:'))
width = float(input('Enter the width of the house in metres:'))
height = float(input('Enter the height of the house in metres:'))
price_per_sq_metre = float(input('Enter the price per square metre of paint in rupees:'))
n_houses = int(input('Enter the number of houses to paint:'))

tpsa, cost = compute_area_cost(length, width, height, price_per_sq_metre, n_houses)

print()
print('Total non-floor surface area of the group of houses:', tpsa, 'square metres')
print('Cost to paint', n_houses, 'houses: ₹' + str(cost))
```
