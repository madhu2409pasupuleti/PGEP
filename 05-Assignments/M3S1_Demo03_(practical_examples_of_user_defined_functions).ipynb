We have a website which sells only one kind of good, a speciality gaming laptop called the PingFortress. Let's design a function which will calculate the price to be paid by the customer ordering a certain quantity of this laptop. Some things to consider are:

*   Each unit of the PingFortress has a volume of 400 cubic centimetres.
*   The function must accept an argument for the quantity of laptops ordered.
*   The function must accept an argument for the price of each laptop.
*   There is a discount offered if more than 5 units of the PingFortress are ordered. The discount is applied before considering shipping charges. Our function must accept an argument for the discount rate.
*   Shipping is charged volumetrically by crate, at a rate of ₹1000 per crate. Each crate can store a volume of 1200 cubic centimetres of productp. The full shipping cost is incurred for a crate even if it isn't full.

```python
def calculate_final_cost(price_per_item, quantity, discount_rate):
    # Fixed parameters
    volume = 400  # volume of each laptop in square centimetres
    volume_per_crate = 1200  # volume of each shipping crate in cubic centimetres
    shipping_cost = 1000  # shipping cost in rupees for each crate
    discount_threshold = 5  # threshold above which a discount is offered, in terms of number of laptops

    # Calculate the total cost before discount and shipping
    total_cost = price_per_item * quantity

    # Calculate the total volume of goods ordered
    total_volume = volume * quantity

    # Apply a discount if quantity is above the discount threshold
    if quantity > discount_threshold:
        total_cost -= total_cost * discount_rate

    # Calculate the number of crates required to ship the goods
    n_crates = total_volume // volume_per_crate  # count the completely filled crates
    if total_volume % volume_per_crate > 0:  # any leftover amount still takes up a crate
        n_crates += 1

    # Add shipping charges
    total_cost += n_crates * shipping_cost

    return total_cost
```

Let's try out our function

```python
calculate_final_cost(price_per_item = 90000, quantity = 10, discount_rate = 0.05)
```

```python
calculate_final_cost(price_per_item = 110000, quantity = 20, discount_rate = 0.06)
```

Let's create a function which checks if a provided year is a leap year or not. A year is considered a leap year if:

*   It is divisible by 4, and
*   If it is divisible by 100, it should also divisible by 400

For instance, 2016 is a leap year as it is divisible by 4 but not 100. 2000 is also a leap year, as it is divisible by 4, 100, and 400. 1900 is not a leap year as although it is divisible by 4, it is also divisible by 100 but not 400.



```python
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True  # divisible by 400, so it's a leap year
            else:
                return False  # divisible by 100 but not 400, so not a leap year
        else:
            return True  # divisible by 4 but not by 100, so it's a leap year
    else:
        return False  # not divisible by 4, so not a leap year
```

Let's try out this function

```python
is_leap_year(1996)
```

```python
is_leap_year(4000)
```

```python
is_leap_year(2100)
```
